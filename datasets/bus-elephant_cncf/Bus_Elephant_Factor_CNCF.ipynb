```python
import json
import urllib.request
import pandas as pd
```

```python
landscape_file = urllib.request.urlopen("https://raw.githubusercontent.com/cncf/tag-contributor-strategy/main/website/data/projects.json")
landscape = json.load(landscape_file)

busDF = pd.read_csv('/Users/dawn/Downloads/All CNCF Users PRs Last year Elephant_Bus Factor stats-data-2023-08-22 10_00_07.csv')
elephantDF = pd.read_csv('/Users/dawn/Downloads/All CNCF Organizations PRs Last year Elephant_Bus Factor stats-data-2023-08-22 10_06_14.csv')

levels_dict = {}
for project in landscape:
    levels_dict[project['name']] = project['project']
    
busDF['level'] = busDF['Project/Repository Group'].map(levels_dict)
busDF.to_csv('bus_by_level.csv')

elephantDF['level'] = elephantDF['Project/Repository Group'].map(levels_dict)
elephantDF.to_csv('elephant_by_level.csv')
```
